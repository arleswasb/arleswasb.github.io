```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
  return 31-((9.8*x)/13)*(1.0 - np.exp(-6.0*(13.0/x)))

def bisseccao(a, b, erro, itMax):
  it = 0
  x = a
  Er = 1
  while(Er >= erro and it < itMax ):
      xOld = x
      x = (a+b)/2
      Er = np.abs((x - xOld)/2)
      if (f(a)*f(x) < 0):
        b = x
      else:
        a = x
      it = it +1
  return(x, Er, it)

a = 40
b = 60

m = np.arange(a,b,1)
plt.plot(m, f(m))
plt.show()

a = 52
b = 55
erro = 10**-4
itMax= 6

res= bisseccao(a , b, erro, itMax)

print('O valor da raíz é:', res[0])
print('O erro relativo foi:', res[1])
print('O numero de iterações realizadas foi:', res[2])
```
